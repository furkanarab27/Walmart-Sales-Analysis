## 5️⃣ Data Visualization


Q1. Weekly sales trend (Line Plot)

```python
sales_trend = df.groupby('Date')['Weekly_Sales'].sum()
plt.figure(figsize=(12,5))
plt.plot(sales_trend.index, sales_trend.values)
plt.title('Weekly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.xticks(rotation=45)
plt.savefig(
    "Weekly)Sales Trend.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

Q2. Monthly sales (Bar Plot)

```python
monthly_trends=df.groupby('Month_Number')['Weekly_Sales'].sum()
plt.figure(figsize=(12,6))
plt.bar(monthly_trends.index,monthly_trends.values)
plt.xticks(
    ticks=monthly_trends.index,
    labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
)
plt.savefig(
    "MOnthly_Sales Trend.png",
    dpi=300,
    bbox_inches="tight"
)
plt.xlabel("Months")
plt.title('Monthly Sales Trends')
plt.ylabel('Sales')
```

Q3.Top 10 stores by sales (Bar Plot)

```python
top10_store=df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
bars=plt.barh(top10_store.index.astype(str),top10_store.values)
plt.ticklabel_format(axis='x',style='plain')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.title('Top 10 Stores by Sales')
plt.xlabel('Sales')
plt.ylabel("Store")
plt.show()
```

Q4. Holiday vs Non-Holiday sales (Bar Plot)

```python
Holiday_Flag=df.groupby('Holiday_Flag')['Weekly_Sales'].sum()
plt.figure(figsize=(10,6))
plt.bar(Holiday_Flag.index,Holiday_Flag.values)
plt.ticklabel_format(style='plain', axis='y')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks([0,1], ['Non-Holiday','Holiday'])
plt.title('Holiday vs Non-Holiday sales')
plt.xlabel('Holiday or Non-Holiday')
plt.ylabel('Sales')
plt.savefig(
    "Holiday_vs_Non Holiday Sales.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

Q5. Temperature distribution (Histogram)

```python
plt.figure(figsize=(10,6))
plt.hist(df['Temperature'],bins=20)
plt.title('Temperature distribution')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Q6. Weekly Sales Distribution

```python
plt.figure(figsize=(10,6))
plt.hist(df['Weekly_Sales'],bins=20)
plt.title('Weekly Sales Distribution')
plt.xlabel('Weekly_Sales')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

Q7. Fuel Price trend (Line Plot)

```python
fuel_trend=df.groupby('Date')['Fuel_Price'].mean()
plt.figure(figsize=(10,6))
plt.plot(fuel_trend.index,fuel_trend.values)
plt.title("Average Fuel Price Over Time")
plt.xlabel('Date')
plt.ylabel('Average Fuel Price')
plt.savefig(
    "Average_Fuel_Price.png",
    dpi=300,
    bbox_inches="tight"
)
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.show()
```

# 6️⃣ Business Insights

- Walmart generated a total sales revenue of **6.74 billion** during the analysis period.

- **Store 20** recorded the highest total sales (**301.40 million**), making it the best-performing store.

- **Store 33** generated the lowest total sales (**37.16 million**) during the analysis period.

- **July** recorded the highest monthly sales among all months.

- **Non-Holiday weeks** generated significantly higher total sales than Holiday weeks because most of the weeks in the dataset are non-holiday weeks.

- The weekly sales trend fluctuated throughout the analysis period, indicating seasonal changes in customer demand.

- The average fuel price changed gradually over time without any sudden spikes.

- Correlation analysis showed that **Temperature, Fuel Price, CPI, and Unemployment have weak correlations with Weekly Sales**, suggesting these variables had little linear relationship with sales in this dataset.
